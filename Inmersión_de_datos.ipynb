```python
inmuebles.groupby('Barrio').mean()
```

```python
#agrupar
datos_barrio = inmuebles.groupby('Barrio').sum()
datos_barrio
```

```python
# mejor opcion para promedios es hacer la suma como el paso anterior y luego dividir por la totalidad
datos_barrio['Valor_m2_Barrio'] = datos_barrio['Precio_Millon']/datos_barrio['Area']
datos_barrio
```

```python
# se crea un dicionario que toma como key el barrio y valor el m2 
m2_barrio = dict(datos_barrio['Valor_m2_Barrio'])
```

```python
# agregamos el dicionario a inmuebles con el valor mas preciso 
inmuebles['Valor_m2_Barrio'] = inmuebles['Barrio']
inmuebles['Valor_m2_Barrio'] = inmuebles['Valor_m2_Barrio'].map(m2_barrio)
inmuebles.head(5)
```

```python
# top 10 
top_barrios = inmuebles['Barrio'].value_counts()[:10].index
```

```python
# para sacar el barrio del indice //solo ejecutar una vez para que no se saque el indice del indice
datos_barrio.reset_index(inplace=True)
datos_barrio
```
